#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esse arquivo.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# %matplotlib inline
%matplotlib notebook

df = pd.read_csv('sinasc_RO_2019.csv')
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
limites = [-1, 3, 5, 7, 10]
rotulos = ['asfixia grave', 'asfixia moderada', 'asfixia leve', 'normal']
df['APGAR5_categorizado'] = pd.cut(df['APGAR5'], bins=limites, labels=rotulos)

print(df[['APGAR5', 'APGAR5_categorizado']].sample(10))
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
plt.figure(figsize=(5, 4))
sns.scatterplot(x='APGAR5', y='GESTACAO', hue='SEXO', data=df, palette='coolwarm')
plt.title('Distribuição de APGAR5 por sexo para cada grupo de tempo de gestação')
plt.xlabel('APGAR5')
plt.ylabel('Tempo de Gestação')
plt.legend(title='Sexo')
plt.tight_layout()
plt.show()
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
tabela = pd.crosstab(index=df['GESTACAO'], columns=df['APGAR5'])
tabela
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
percentuais = tabela.div(tabela.sum(axis=1), axis=0) * 100
percentuais
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
plt.figure(figsize=(10, 8))
percentuais.plot(kind='bar', width=1)
plt.title('Distribuição de APGAR5 por classe de GESTACAO')
plt.xlabel('Classe de GESTACAO')
plt.ylabel('Porcentagem')
plt.legend(title='APGAR5', bbox_to_anchor=(1,1.05), loc='upper left')
plt.tight_layout()
plt.xticks(rotation=18)
plt.show()
```

Com base no gráfico fornecido, verifica-se que esse conjunto de dados corrobora a afirmação dos médicos. A distribuição do APGAR5 nos diferentes períodos gestacionais demonstra uma clara tendência: à medida que a duração gestacional diminui, aumenta a proporção de crianças com menores APGAR5 (maior asfixia), indicando potenciais complicações ao nascer.
